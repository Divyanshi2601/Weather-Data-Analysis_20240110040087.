Weather Data Analysis â€“ temperature trends, rainfall, and seasonal patterns from a small weather dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the weather dataset
data = {
    "Date": pd.date_range(start="2024-01-05", periods=12, freq='M'),
    "Temperature (Â°C)": [12.5, 14.2, 18.4, 22.8, 27.3, 30.5, 33.8, 32.1, 26.4, 20.3, 15.7, 10.9],
    "Rainfall (mm)": [5.2, 2.8, 1.5, 0.0, 3.1, 10.4, 15.2, 12.7, 8.3, 6.1, 4.2, 7.5],
    "Humidity (%)": [75, 72, 68, 60, 55, 50, 52, 53, 58, 65, 70, 74],
    "Wind Speed (km/h)": [10, 12, 14, 16, 18, 20, 22, 21, 19, 15, 13, 11],
    "Season": ["Winter", "Winter", "Spring", "Spring", "Summer", "Summer", "Summer", "Summer", "Fall", "Fall", "Fall", "Winter"]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert Date column to datetime format
df["Date"] = pd.to_datetime(df["Date"])

# Set Date as the index for plotting
df.set_index("Date", inplace=True)

# Plot 1: Temperature Trend Over Time
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["Temperature (Â°C)"], marker='o', linestyle='-', color='r', label="Temperature (Â°C)")
plt.xlabel("Date")
plt.ylabel("Temperature (Â°C)")
plt.title("Temperature Trends Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()

# Plot 2: Rainfall Distribution
plt.figure(figsize=(10, 5))
plt.bar(df.index, df["Rainfall (mm)"], color='blue', alpha=0.7, label="Rainfall (mm)")
plt.xlabel("Date")
plt.ylabel("Rainfall (mm)")
plt.title("Rainfall Distribution Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.show()

# Plot 3: Seasonal Temperature Variations (Box Plot)
plt.figure(figsize=(8, 5))
df.boxplot(column="Temperature (Â°C)", by="Season", grid=False, patch_artist=True)
plt.title("Seasonal Temperature Variations")
plt.suptitle("")  # Remove default title
plt.xlabel("Season")
plt.ylabel("Temperature (Â°C)")
plt.show()
```
